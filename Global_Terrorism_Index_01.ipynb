## Fine-tune random forest regressor hyperparameters

### Subtask:
Fine-tune the hyperparameters of the Random Forest Regressor using techniques like GridSearchCV or RandomizedSearchCV to potentially improve its performance further.


**Reasoning**:
Fine-tune the hyperparameters of the Random Forest Regressor using RandomizedSearchCV to potentially improve its performance further by defining a parameter distribution, instantiating and fitting RandomizedSearchCV, printing the best parameters, instantiating and training a new model with the best parameters, making predictions, and evaluating the tuned model's performance.



```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Define the parameter distribution
param_dist = {'n_estimators': np.arange(100, 500, 100),
              'max_depth': [None, 10, 20, 30],
              'min_samples_split': np.arange(2, 10, 2),
              'min_samples_leaf': np.arange(1, 5, 1),
              'bootstrap': [True, False]}

# Instantiate RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator=RandomForestRegressor(random_state=42),
                               param_distributions=param_dist,
                               n_iter=10, # Number of parameter settings that are sampled
                               cv=5, # Number of cross-validation folds
                               random_state=42,
                               n_jobs=-1) # Use all available cores

# Fit RandomizedSearchCV to the training data
rf_random.fit(X_train, y_train)

# Print the best parameters
print("Best parameters found by RandomizedSearchCV:")
display(rf_random.best_params_)

# Instantiate a new RandomForestRegressor model with the best parameters
best_rf_model = RandomForestRegressor(**rf_random.best_params_, random_state=42)

# Train the new Random Forest model with the best parameters on the entire training data
best_rf_model.fit(X_train, y_train)

# Make predictions on the test set using the tuned model
y_pred_tuned_rf = best_rf_model.predict(X_test)

# Evaluate the tuned model's performance
mse_tuned_rf = mean_squared_error(y_test, y_pred_tuned_rf)
r2_tuned_rf = r2_score(y_test, y_pred_tuned_rf)

# Print the evaluation metrics
print("\nTuned Random Forest Regressor Performance:")
print(f"  Mean Squared Error: {mse_tuned_rf}")
print(f"  R-squared: {r2_tuned_rf}")
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Define the ARIMA order (p, d, q) - based on ACF/PACF analysis and differencing
# Using (1, 1, 1) as a starting point based on the plots
order = (1, 1, 1)

# Instantiate and train the ARIMA model
arima_model = ARIMA(incidents_ts, order=order)
arima_results = arima_model.fit()

# Print the model summary
display(arima_results.summary())
```

## Identify ARIMA parameters

### Subtask:
Determine the appropriate p, d, and q parameters for the ARIMA model using techniques like ACF and PACF plots or auto_arima.

**Reasoning**:
Since the ADF test indicated non-stationarity, apply first-order differencing to make the time series stationary, then plot ACF and PACF to identify potential p and q values for the ARIMA model.

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Apply first-order differencing
incidents_ts_diff = incidents_ts.diff().dropna()

# Plot ACF and PACF of the differenced series
fig, axes = plt.subplots(2, 1, figsize=(10, 8))
plot_acf(incidents_ts_diff, ax=axes[0])
plot_pacf(incidents_ts_diff, ax=axes[1])
plt.show()
```

## Check for stationarity

### Subtask:
Perform tests (e.g., Augmented Dickey-Fuller test) to check if the time series is stationary. If not, apply differencing.

**Reasoning**:
Perform the Augmented Dickey-Fuller test to check for stationarity and interpret the p-value.

```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
result = adfuller(incidents_ts)

# Print the results
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

# Interpret the results
if result[1] <= 0.05:
    print("\nInterpretation: The time series is likely stationary.")
else:
    print("\nInterpretation: The time series is likely non-stationary. Differencing may be needed.")
```

Certainly! The Random Forest Regressor has several hyperparameters that control its behavior and performance. In the RandomizedSearchCV step, we experimented with the following hyperparameters:

min_samples_split: This is the minimum number of samples required to split an internal node. A higher value prevents the tree from splitting on small groups of samples, which can help reduce overfitting. We explored values from 2 to 8 in steps of 2.
min_samples_leaf: This is the minimum number of samples required to be at a leaf node. We explored values from 1 to 4 in steps of 1.
bootstrap: This is a boolean hyperparameter that determines whether bootstrap samples are used when building trees. If True, each tree is built on a bootstrap sample of the data. If False, the entire dataset is used to build each tree. Bootstrapping can help reduce variance. We explored both True and False.
RandomizedSearchCV randomly samples combinations of these hyperparameters from the specified distributions and evaluates the model's performance for each combination using cross-validation. This helps find a good set of hyperparameters without exhaustively searching through all possible combinations. ## Define the objective

### Subtask:
Clearly define the objective of the k-means clustering (e.g., group countries based on terrorism-related factors).

State the objective of the k-means clustering analysis.



```python
# Objective:
# The primary objective of this k-means clustering analysis is to group countries based on their terrorism-related factors, such as incidents, fatalities, injuries, and hostages.
# This clustering aims to identify countries with similar patterns of terrorism activity, which can help in understanding regional differences and tailoring counter-terrorism strategies.
```

## Prepare the data

### Subtask:
Select the relevant numerical columns for clustering and handle any missing values or outliers.


**Reasoning**:
Create a new DataFrame with the relevant numerical columns for clustering and display the head and info.



```python
clustering_data = df[['Score', 'Incidents', 'Fatalities', 'Injuries', 'Hostages']]
display(clustering_data.head())
display(clustering_data.info())
```

**Reasoning**:
Check for and handle any missing values in the selected columns and display the sum of null values per column.



```python
display(clustering_data.isnull().sum())
# As seen from the previous info output, there are no missing values.
# No outlier handling is applied at this stage, as it can be sensitive to the clustering algorithm and might distort the data distribution.
```

## Standardize the data

### Subtask:
Standardize the numerical features to ensure they have a similar scale.


**Reasoning**:
Standardize the numerical features using StandardScaler.



```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data)
clustering_data_scaled = pd.DataFrame(clustering_data_scaled, columns=clustering_data.columns)
display(clustering_data_scaled.head())
```

## Determine the optimal number of clusters (k)

### Subtask:
Determine the optimal number of clusters (k) using the elbow method.


**Reasoning**:
Determine the optimal number of clusters (k) using the elbow method by iterating through a range of possible cluster numbers, fitting KMeans for each, and storing the inertia values, then plotting the inertia values against the number of clusters.



```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42, n_init=10) # Added n_init to suppress warning
    kmeans.fit(clustering_data_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

## Apply k-means clustering

### Subtask:
Apply the k-means algorithm with the chosen number of clusters to the standardized data.


**Reasoning**:
Apply the k-means algorithm with the chosen number of clusters to the standardized data, predict the cluster labels, and add them to the original dataframe.



```python
from sklearn.cluster import KMeans

# Based on the elbow method plot, let's choose k=3 as a starting point
n_clusters = 3

kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
kmeans.fit(clustering_data_scaled)
df['Cluster'] = kmeans.predict(clustering_data_scaled)

display(df.head())
```
