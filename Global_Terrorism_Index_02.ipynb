## Perform sensitivity analysis

### Subtask:
Rerun the clustering with different numbers of clusters (k) or different random initializations to assess the stability of the resulting clusters.

**Reasoning**:
Iterate through different numbers of clusters and random states, fit KMeans, calculate silhouette score and Davies-Bouldin index, and print the results.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import pandas as pd

# Define a range of cluster numbers and random states to test
n_clusters_range = [2, 3, 4, 5]
random_states = [0, 42, 100]

results = []

for n_clusters in n_clusters_range:
    for random_state in random_states:
        print(f"\nTesting n_clusters={n_clusters}, random_state={random_state}")

        # Instantiate and fit KMeans
        kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
        kmeans.fit(clustering_data_scaled)
        cluster_labels = kmeans.predict(clustering_data_scaled)

        # Calculate evaluation metrics
        try:
            silhouette_avg = silhouette_score(clustering_data_scaled, cluster_labels)
            db_index = davies_bouldin_score(clustering_data_scaled, cluster_labels)

            print(f"  Silhouette Score: {silhouette_avg:.4f}")
            print(f"  Davies-Bouldin Index: {db_index:.4f}")

            results.append({
                'n_clusters': n_clusters,
                'random_state': random_state,
                'silhouette_score': silhouette_avg,
                'davies_bouldin_index': db_index
            })
        except ValueError as e:
            print(f"  Could not calculate metrics: {e}")
            results.append({
                'n_clusters': n_clusters,
                'random_state': random_state,
                'silhouette_score': None,
                'davies_bouldin_index': None
            })

# Display results in a DataFrame for easier comparison
results_df = pd.DataFrame(results)
print("\nSummary of Clustering Robustness Tests:")
display(results_df)
```

Explain the hyperparameters of the Random Forest Regressor
Certainly! The Random Forest Regressor has several hyperparameters that control its behavior and performance. In the RandomizedSearchCV step, we experimented with the following hyperparameters:

n_estimators: This is the number of trees in the forest. A higher number of trees generally leads to better performance but also increases computation time. We explored values from 100 to 400 in steps of 100.
max_depth: This is the maximum depth of each tree in the forest. It controls how deeply each tree is allowed to grow. None means that nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. Limiting the depth can help prevent overfitting. We explored None, 10, 20, and 30.
min_samples_split: This is the minimum number of samples required to split an internal node. A higher value prevents the tree from splitting on small groups of samples, which can help reduce overfitting. We explored values from 2 to 8 in steps of 2.
min_samples_leaf: This is the minimum number of samples required to be at a leaf node. A split point at any depth is only considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. This can also help prevent overfitting. We explored values from 1 to 4 in steps of 1.
bootstrap: This is a boolean hyperparameter that determines whether bootstrap samples are used when building trees. If True, each tree is built on a bootstrap sample of the data. If False, the entire dataset is used to build each tree. Bootstrapping can help reduce variance. We explored both True and False.
RandomizedSearchCV randomly samples combinations of these hyperparameters from the specified distributions and evaluates the model's performance for each combination using cross-validation. This helps find a good set of hyperparameters without exhaustively searching through all possible combinations. The best parameters found by RandomizedSearchCV were then used to train the final tuned Random Forest model.

## Perform k-means clustering

### Subtask:
Apply the k-means algorithm with the chosen number of clusters to the standardized data and add the cluster labels to the dataframe.

**Reasoning**:
Apply the k-means algorithm with the chosen number of clusters to the standardized data, predict the cluster labels, and add them to the original dataframe.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Assuming 'clustering_data' and 'n_clusters' are already defined from previous steps
# If not, you would need to include those steps here.
# For this example, let's assume we are using the same data and k=3 as before.

# Select the relevant numerical columns for clustering
clustering_data = df[['Score', 'Incidents', 'Fatalities', 'Injuries', 'Hostages']]

# Standardize the data
scaler = StandardScaler()
clustering_data_scaled = scaler.fit_transform(clustering_data)
clustering_data_scaled = pd.DataFrame(clustering_data_scaled, columns=clustering_data.columns)

# Apply k-means clustering (assuming k=3 based on previous analysis)
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
kmeans.fit(clustering_data_scaled)
df['Cluster'] = kmeans.predict(clustering_data_scaled)

display(df.head())
```

## Perform sensitivity analysis

### Subtask:
Rerun the clustering with different numbers of clusters (k) or different random initializations to assess the stability of the resulting clusters.


**Reasoning**:
Iterate through different numbers of clusters and random states, fit KMeans, calculate silhouette score and Davies-Bouldin index, and print the results.



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
import pandas as pd

# Define a range of cluster numbers and random states to test
n_clusters_range = [2, 3, 4, 5]
random_states = [0, 42, 100]

results = []

for n_clusters in n_clusters_range:
    for random_state in random_states:
        print(f"\nTesting n_clusters={n_clusters}, random_state={random_state}")

        # Instantiate and fit KMeans
        kmeans = KMeans(n_clusters=n_clusters, random_state=random_state, n_init=10)
        kmeans.fit(clustering_data_scaled)
        cluster_labels = kmeans.predict(clustering_data_scaled)

        # Calculate evaluation metrics
        try:
            silhouette_avg = silhouette_score(clustering_data_scaled, cluster_labels)
            db_index = davies_bouldin_score(clustering_data_scaled, cluster_labels)

            print(f"  Silhouette Score: {silhouette_avg:.4f}")
            print(f"  Davies-Bouldin Index: {db_index:.4f}")

            results.append({
                'n_clusters': n_clusters,
                'random_state': random_state,
                'silhouette_score': silhouette_avg,
                'davies_bouldin_index': db_index
            })
        except ValueError as e:
            print(f"  Could not calculate metrics: {e}")
            results.append({
                'n_clusters': n_clusters,
                'random_state': random_state,
                'silhouette_score': None,
                'davies_bouldin_index': None
            })

# Display results in a DataFrame for easier comparison
results_df = pd.DataFrame(results)
print("\nSummary of Clustering Robustness Tests:")
display(results_df)
```
